```python
import pandas as pd

# Load dataset
df = pd.read_csv("final_rag.csv")

# 1️⃣ Create branch short forms mapping
branch_map = {
    "COMPUTER SCIENCE AND ENGINEERING": "CSE",
    "INFORMATION SCIENCE AND ENGINEERING": "ISE",
    "ELECTRONICS AND COMMUNICATION ENGINEERING": "ECE",
    "ELECTRICAL AND ELECTRONICS ENGINEERING": "EEE",
    "MECHANICAL ENGINEERING": "ME",
    "CIVIL ENGINEERING": "CE",
    "ARTIFICIAL INTELLIGENCE AND DATA SCIENCE": "AI & DS",
    "DATA SCIENCE": "DS",
    "ROBOTICS AND AUTOMATION ENGINEERING": "RAE",
    "AEROSPACE ENGINEERING": "AE",
    "CHEMICAL ENGINEERING": "CHE",
    # add more as needed
}

# 2️⃣ Add the Branch_Short column
df['Branch_Short'] = df['Branch'].map(branch_map).fillna(df['Branch'])

# 3️⃣ Combine fields into a single string for embeddings
df['content'] = (
    df['College'].astype(str) + " | " +
    df['Branch'].astype(str) + " (" + df['Branch_Short'].astype(str) + ")" + " | " +
    df['Category'].astype(str) + " | Cutoff: " + df['Cutoff_rank'].astype(str) +
    " | Exam: " + df['Exam'].astype(str) +
    " | Year: " + df['Year'].astype(str) +
    " | Tuition: " + df['Tuition_Fee'].astype(str) +
    " | Hostel: " + df['Hostel_Fee'].astype(str) +
    " | Total_Annual: " + df['Total_Annual'].astype(str) +
    " | Avg_Package: " + df['Avg_Package_LPA'].astype(str) +
    " | Top_Companies: " + df['Top_Companies'].astype(str)
)

# 4️⃣ Check sample
print(df['content'].head(3))
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
import pickle



model_name = "all-mpnet-base-v2"  # high-quality embeddings
model = SentenceTransformer(model_name)


contents = df['content'].tolist()
embeddings = model.encode(contents, batch_size=64, show_progress_bar=True)
embeddings = np.array(embeddings).astype('float32')
print("Embeddings shape:", embeddings.shape)
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)  # L2 distance
index.add(embeddings)

print("Number of vectors in FAISS index:", index.ntotal)
```

```python
# Save FAISS index
faiss.write_index(index, "final_rag_index.faiss")

# Save the DataFrame for later retrieval
df.to_pickle("final_rag_df.pkl")
```

```python
queries = [
    "Best colleges for COMPUTER SCIENCE AND ENGINEERING (CSE) under KCET rank 6000",
    "Top ELECTRONICS AND COMMUNICATION ENGINEERING (ECE) colleges with OBC category cutoff below 5000",
    "Affordable DATA SCIENCE (DS) programs with high placement packages",
    "Artificial Intelligence and Data Science (AI & DS) courses for GM category under 7000 rank",
    "MECHANICAL ENGINEERING (ME) colleges with avg package above 8 LPA",
    "ROBOTICS AND AUTOMATION ENGINEERING (RAE) colleges open for SC category students",
    "Good CIVIL ENGINEERING (CE) colleges under COMEDK rank 10000",
    "Best ELECTRICAL AND ELECTRONICS ENGINEERING (EEE) colleges with hostel facilities",
    "Top AEROSPACE ENGINEERING (AE) colleges under KCET 5000",
    "Colleges offering INFORMATION SCIENCE AND ENGINEERING (ISE) for GM category with high placements"
]
```

```python
k = 5

for query in queries:
    query_emb = model.encode([query]).astype('float32')
    distances, indices = index.search(query_emb, k)
    print(f"\nQuery: {query}")
    for i, idx in enumerate(indices[0]):
        print(f"Rank {i+1}: {df.iloc[idx]['content']} (Distance: {distances[0][i]:.2f})")
```

```python
query = "Best colleges for COMPUTER SCIENCE AND ENGINEERING (CSE) under KCET rank 6000"
query_emb = model.encode([query]).astype('float32')

k = 20  # Retrieve more than needed to allow post-filtering
distances, indices = index.search(query_emb, k)
results = df.iloc[indices[0]].copy()
results['distance'] = distances[0]
```

```python
# Only colleges with KCET rank <= 6000
results_filtered = results[results['Cutoff_rank'] <= 6000]

# Optional: filter by minimum Avg_Package
results_filtered = results_filtered[results_filtered['Avg_Package_LPA'] >= 5.0]
```

```python
# Keep the best match per College+Branch combination (lowest distance)
results_dedup = results_filtered.sort_values('distance').drop_duplicates(subset=['College', 'Branch'], keep='first')
```

```python
top_results = results_dedup.head(5)
for i, row in enumerate(top_results.itertuples(), 1):
    print(f"Rank {i}: {row.content} (Distance: {row.distance:.2f})")
```
